# Dataset

KC Crime Datasets from 2014-2018

https://data.kcmo.org/Crime/KCPD-Crime-Data-2015/kbzx-7ehe

```python
import pandas as pd
import io
import requests

# 2014 Dataset 
url="https://data.kcmo.org/resource/nsn9-g8a4.csv"
s=requests.get(url).content
df_2014=pd.read_csv(io.StringIO(s.decode('utf-8')))

# 2015 Dataset 
url="https://data.kcmo.org/resource/geta-wrqs.csv"
s=requests.get(url).content
df_2015=pd.read_csv(io.StringIO(s.decode('utf-8')))


# 2016 Dataset  
url="https://data.kcmo.org/resource/c6e8-258d.csv"
s=requests.get(url).content
df_2016=pd.read_csv(io.StringIO(s.decode('utf-8')))

# 2017 Dataset 
url="https://data.kcmo.org/resource/wy8a-bydn.csv"
s=requests.get(url).content
df_2017=pd.read_csv(io.StringIO(s.decode('utf-8')))

# 2018 Dataset 
url="https://data.kcmo.org/resource/nyg5-tzkz.csv"
s=requests.get(url).content
df_2018=pd.read_csv(io.StringIO(s.decode('utf-8')))

df_2014['Year'] = '2014'
df_2015['Year'] = '2015'
df_2016['Year'] = '2016'
df_2017['Year'] = '2017'
df_2018['Year'] = '2018'
```

```python
df = df_2014
df.append(df_2015)
df.append(df_2016)
df.append(df_2017)
df.append(df_2018)
pd.options.display.max_columns = 999
df.head()
```
